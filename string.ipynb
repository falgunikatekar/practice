```python
#1. Reverse a String
def rev(s):
  print(s[::-1])
rev("hello")
```

```python
#Check Palindrome
def pal(s):
  r1=s
  r2=s[::-1]
  if(r1==r2):
    print("palindrom")
  else:
    print("not palindrom")

pal("mdam")
```

```python
def dup(s):
    result = ""
    for x in s:
        if x not in result:   # only add if not already included
            result += x
    return result

print(dup("apple"))       # output: 'aple'
print(dup("programming")) # output: 'progamin'
```

```python
# check if a string contains only digits
def nocheck(s):
    if s.isdigit():
        print("string contains only digits")
    else:
        print("string has non-digit characters")

nocheck("1245")       # ✅ string contains only digits
nocheck("12a45")      # ❌ string has non-digit characters
```

```python
#count a number of vowels and consonants in a given string
def count_(s):
  vowel_count = 0
  consonant_count = 0
  for ch in s:
    if 'a' <= ch.lower() <= 'z': # Check if it's an alphabet character
        if ch.lower() in 'aeiou':
            vowel_count += 1
        else:
            consonant_count += 1
  return f"Vowels: {vowel_count}, Consonants: {consonant_count}"

print(count_("apple"))
print(count_("programming"))
```

```python
#First Non-Repeating Character
#Input: "swiss" → Output: "w"
def first_non_repeating(s):
    for ch in s:
        if s.count(ch) == 1:   # check if character appears only once
            return ch
    return None   # if no non-repeating character exists

print(first_non_repeating("swiss"))   # Output: w
print(first_non_repeating("apple"))   # Output: a
print(first_non_repeating("programming")) # Output: o

```

```python
#Count Occurrences of Each Character
def count(s):
  char_counts = {}
  for ch in s:
    char_counts[ch] = char_counts.get(ch, 0) + 1
  return char_counts

print(count("apple"))
print(count("programming"))
```

```python
#Longest Substring Without Repeating Characters
#Input: "abcabcbb" → Output: 3 ("abc")
def longest(s):
    seen = {}
    start = 0
    max_len = 0
    substring = ""

    for i, ch in enumerate(s):
        if ch in seen and seen[ch] >= start:
            start = seen[ch] + 1
        seen[ch] = i
        if i - start + 1 > max_len:
            max_len = i - start + 1
            substring = s[start:i+1]

    return max_len, substring

print(longest("abcabcbb"))  # Output: (3, 'abc')
print(longest("bbbbb"))     # Output: (1, 'b')
print(longest("pwwkew"))    # Output: (3, 'wke')

```
