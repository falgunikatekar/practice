```python
# subarray of arraySample Output Description 1
#Here, N = 11, k = 2
#A = [1, 2, 2, 3, 2, 3, 5, 1, 2, 1, 1]
def subarrays(arr, k):
    result = []

    for i in range(len(arr)):
        total = 0
        temp = []

        for j in range(i, len(arr)):
            total += arr[j]
            temp.append(arr[j])

            if total == k:
                result.append(temp[:])

    return result


arr = [1, 2, 2, 3, 2, 3, 5, 1, 2, 1, 1]
print(subarrays(arr, 12))

```
